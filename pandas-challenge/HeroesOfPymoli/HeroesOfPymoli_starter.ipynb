```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

game_purchases_df["Age Group"] = pd.cut(game_purchases_df["Age"],age_bins, labels=group_names)

```

```python
age_grouping = game_purchases_df.groupby("Age Group")

total_age_group = age_grouping["SN"].nunique()

 
percentage_by_age = (total_age_group/player_count) * 100


age_demo = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": total_age_group})


age_demo.index.name = None

 
age_demo.style.format({"Percentage of Players":"{:,.2f}"})
```

```python
# Count purchases by age group
purchase_count_age = age_grouping["Purchase ID"].count()

# Obtain average purchase price by age group 
avg_purchase_price_age = age_grouping["Price"].mean()

# Calculate total purchase value by age group 
total_purchase_value = age_grouping["Price"].sum()

# Calculate the average purchase per person in the age group 
avg_purchase_per_person_age = total_purchase_value/total_age_group

# Create data frame with obtained values
age_demo =         pd.DataFrame({"Purchase Count": purchase_count_age,
                                 "Average Purchase Price": avg_purchase_price_age,
                                 "Total Purchase Value":total_purchase_value,
                                 "Average Purchase Total per Person": avg_purchase_per_person_age})

age_demo
```
